```python
import matplotlib.pyplot as plt
from PIL import Image

# 关闭 PIL 的最大像素限制（防止 DecompressionBombError）
Image.MAX_IMAGE_PIXELS = None

# 图片路径
img_path = "/mnt/raid/zanzhuheng/working/sheep/heatmap_viz/5.ome_merged_pyramid/heatmap_phikon_v2_AB_MIL_OPA.png"

# 读取图片
img = Image.open(img_path).convert("RGB")

# resize 到 1024x1024
img = img.resize((1024, 1024), Image.Resampling.BICUBIC)

# plt 显示
plt.figure(figsize=(16, 16))  # 仅影响显示尺寸
plt.imshow(img)
plt.axis("off")
plt.title("Heatmap (1024 x 1024)")
plt.show()
```
